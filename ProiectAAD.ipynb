```python
df_merge_lifeexpectency_cost = df_food_prices.iloc[:,0:5]
df_merge_lifeexpectency_cost = df_merge_lifeexpectency_cost.merge(df_important, how="inner", on=['Country', 'Year'])
display(df_merge_lifeexpectency_cost)
df_merge_meat_lifeexpectency = df_merge_lifeexpectency_cost.merge(df_meat_consumption, how="inner", on=['Country', 'Year'])
display(df_merge_meat_lifeexpectency)
df_merge_total = df_merge_meat_lifeexpectency.merge(df_global_price, how="inner", on=['Country', 'Year'])
display(df_merge_total)
df_merge_total = df_merge_total.dropna(axis=1, thresh=df_merge_total.shape[0] * 0.5)
display(df_merge_total.describe())
```

```python
display(df_merge_total.head())
df_merge_final = df_merge_total.replace(np.nan, 0);
display(df_merge_final)

print(df_merge_final.isna().sum())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#df_merge_final = df_merge_final.drop('Country', axis=1)
# df_merge_final = df_merge_final.drop('Gender', axis=1) # sa ii dai encode in viitor

X = df_merge_final.drop('Life Expectancy', axis=1)
Y = df_merge_final['Life Expectancy']

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42)

reg = LinearRegression().fit(X_train, Y_train)
Y_predict = reg.predict(X_test)

print(mean_squared_error(Y_test, Y_predict))
print(r2_score(Y_test, Y_predict))

display(Y_predict[0:5])
display(Y_test.head())
```
